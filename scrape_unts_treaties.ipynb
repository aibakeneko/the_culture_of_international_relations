```python

sys.path.append(os.path.abspath(os.path.join('..')))



# 'https://treaties.un.org/doc/Publication/UNTS/Volume%20263/volume-263-I-3766-Other.pdf'

repository = WorldTreatyIndexRepository()

languages = [ ('English', 'en'), ('French', 'fr'), ('Other', 'other') ]

url_template = 'https://treaties.un.org/doc/Publication/UNTS/Volume%20{0}/{1}'
filename_template = 'volume-{0}-{3}-{1}-{2}.pdf'

result = pd.DataFrame(columns=['treaty_id', 'volume', 'party1', 'party2', 'signed', 'language', 'http_code', 'filename', 'status', 'url'])

for (treaty_id, volume, party1, party2, signed) in treaties:

    for (language, lang) in languages:

        for filename in [ filename_template.format(volume, int(str(treaty_id)[-5:]), language, 'I'),filename_template.format(volume, int(str(treaty_id)[-5:]), language, 'II') ]:

            url = url_template.format(volume, filename)
            status = False
            code = 0
            try:
                response = urllib.request.urlopen(url)
                code = response.getcode()
                status = code in [ 200 ]
                if status:
                    pdf = response.read()
                    savename = "{}-{}.pdf".format(treaty_id, lang)
                    path = repository.data_path(savename)
                    with open(path,'wb') as output:
                        output.write(pdf)
                    break

            except Exception as ex:
                #print(str(ex))
                status = False

        print('{}/{};{};{};{}'.format(volume, treaty_id, language, status, url))

        result = result.append({
                'treaty_id' : treaty_id,
                'volume' : volume,
                'party1': party1,
                'party2': party2,
                'signed': signed,
                'language': language,
                'http_code': code,
                'filename': filename,
                'status': status,
                'url': url
        }, ignore_index=True)

save_path = repository.data_path('UNTS-download-log.xlsx')
repository.save_excel([(result, 'Data')], save_path)
```
